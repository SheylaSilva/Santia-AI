```python
from gtts import gTTS
from IPython.display import Audio, display

# Função para converter texto em fala
def texto_para_fala(texto, lang='pt'):
    tts = gTTS(text=texto, lang=lang)
    tts.save("output.mp3")
    display(Audio("output.mp3", autoplay=True))

# Exemplo de uso
texto = "Olá, este é um teste de conversão de texto para fala."
texto_para_fala(texto)
```

```python
# Instalar Coqui TTS
!pip install TTS

# Importar e usar Coqui TTS
from TTS.api import TTS

# Baixar e carregar modelo
tts = TTS(model_name="tts_models/en/ljspeech/tacotron2-DDC", progress_bar=True, gpu=False)

# Converter texto em fala
tts.tts_to_file(text="Hello, this is a voice cloning test.", file_path="output.wav")
```
